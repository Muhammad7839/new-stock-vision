```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train_df = pd.read_csv('../archive/financial_data/final_training_data.csv')
test_df = pd.read_csv('../archive/financial_data/final_testing_data.csv')
```

```python
print(train_df.shape)
print(train_df.columns)
print(train_df.head())
```

```python
# // step 5 Check for missing values
```

```python
# Check data types
print("Data Types:\n")
print(train_df.dtypes)

# Check for missing values
print("\nMissing Values Per Column:\n")
print(train_df.isnull().sum())
```

```python
# //Step 6  Check for Duplicates
```

```python
# Step 6: Check for duplicate rows in the dataset
duplicate_count = train_df.duplicated().sum()
print("Number of duplicate rows:", duplicate_count)

# Show duplicate rows if any
if duplicate_count > 0:
    display(train_df[train_df.duplicated()])
```

```python
# Check for Outliers
```

```python
from scipy.stats import zscore
import numpy as np

# Select only numeric columns
numeric_cols = train_df.select_dtypes(include=[np.number])

# Calculate z-scores
z_scores = np.abs(zscore(numeric_cols))

# Mark values with z-score > 3 as outliers
outliers = (z_scores > 3)

# Count how many outliers per column
outlier_counts = np.sum(outliers, axis=0)

# Show result as a pandas series
outliers_series = pd.Series(outlier_counts, index=numeric_cols.columns)

# Sort and print top columns with most outliers
print("Top columns with outliers:")
print(outliers_series.sort_values(ascending=False).head(10))
```

```python
#  Remove Low Variance Columns
```

```python
# Step 8: Remove low variance columns

# Calculate standard deviation for each numeric column
std_devs = train_df.select_dtypes(include=[np.number]).std()

# Set a threshold for low variance (you can change this if needed)
low_variance_threshold = 0.001

# Get columns with very low variance
low_variance_cols = std_devs[std_devs < low_variance_threshold].index

# Drop those columns from the training data
train_df = train_df.drop(columns=low_variance_cols)

print("Removed low variance columns:")
print(low_variance_cols)
print("Updated shape of training data:", train_df.shape)
```

```python
# Detect Multicollinearity.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Compute correlation matrix for numeric columns
correlation_matrix = train_df.select_dtypes(include='number').corr()

# Set a threshold to detect multicollinearity
threshold = 0.9

# Create a mask to filter only strong correlations (excluding self-correlations)
mask = (abs(correlation_matrix) > threshold) & (correlation_matrix != 1.0)

# Find column pairs with high correlation
high_corr_pairs = mask.stack()[mask.stack()].index.tolist()

# Display the top correlated column pairs
print("Highly correlated pairs (correlation > 0.9):")
for col1, col2 in high_corr_pairs:
    print(f"{col1}  {col2}  |  corr = {correlation_matrix.loc[col1, col2]:.2f}")

# Optional: visualize correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, cmap='coolwarm', vmax=1.0, vmin=-1.0)
plt.title("Correlation Heatmap")
plt.show()
```
